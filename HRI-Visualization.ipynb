```python
#identify image size:
def find_max_min_XY(all_points):
    xs=[]
    ys=[]
    for x,y in all_points:
        xs.append(float(x))
        ys.append(float(y))
    return max(xs),max(ys),min(xs),min(ys)

def find_min_XY(all_points):
    xs=[]
    ys=[]
    for x,y in all_points:
        xs.append(float(x))
        ys.append(float(y))
    return min(xs),min(ys)

all_points=[]
for room, points in roomdict.items():
    all_points+=points
    
for counter, player_data in playerwise_data.items():
    player_path,robot_path,rooms_visited,c=player_data
    for point in player_path:
        all_points.append(point)
    for point in robot_path:
        all_points.append(point)
        
maxx,maxy,minx,miny = find_max_min_XY(all_points)


w=int(maxx-minx)
h=int(maxy-miny)
w+=20
h+=20
multiplierx=30
multipliery=30
w=w*multiplierx
h=h*multipliery
```

```python
def adjust_path_to_scale(path,modifierX,modifierY, scaleX,scaleY):
    newpath=[]

    for x,y in path:
        newpath.append(((x+modifierX)*scaleX,(y+modifierY)*scaleY))

    return newpath

def adjust_point_to_scale(point,modifierX,modifierY, scaleX,scaleY):
    x,y=point
    return((x+modifierX)*scaleX,(y+modifierY)*scaleY)
```

```python
#Fuctions for drawing paths:
def line_dashed(point0, point1, dashlen=4, ratio=2, colour=(0,0,0), width=2): 
    x0=point0[0]
    y0=point0[1]
    x1=point1[0]
    y1=point1[1]
    dx=x1-x0 # delta x
    dy=y1-y0 # delta y
    # check whether we can avoid sqrt
    if dy==0: 
        length=dx
    elif dx==0: 
        length=dy
    else: 
        length=math.sqrt(dx*dx+dy*dy) # length of line
        xa=dx/length # x add for 1px line length
        ya=dy/length # y add for 1px line length
        step=dashlen*ratio # step to the next dash
        a0=0
        while a0<length:
            a1=a0+dashlen
            if a1>length: a1=length
            draw.line((x0+xa*a0, y0+ya*a0, x0+xa*a1, y0+ya*a1), fill = colour, width= width)
            a0+=step 
            
def draw_path(path, style="regular", colour="#0000ff", width=5):
    #adjust points in player_path to all be positive
    path=adjust_path_to_scale(path,modifierx,modifiery, multiplierx,multipliery)

    #draw the player_path
    for i in range(0,len(path)-1):
        if style=="regular":
            draw.line(path[i]+path[i+1], fill=colour, width=width)
        elif style=="dashed":
            line_dashed(path[i], path[i+1], dashlen=5, ratio=6, colour=colour, width=width)

def draw_point(point, color):
    point=adjust_point_to_scale(point,modifierx,modifiery, multiplierx,multipliery)
    draw.regular_polygon((point,40),6, fill=color)
```
